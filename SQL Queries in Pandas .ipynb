### We join multiple conditions with &. If we only want a subset of columns from the table, that subset is applied in another pair of square brackets.

```python
df[(df["gender"] == "Male") & (df["Partner"] == "Yes")][['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure']]
```

```python
df[(df.gender == "Male") & (df.Partner == "Yes")][['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure']]
```

### Order by : By default, Pandas will sort things in ascending order. To reverse that, provide ascending=False.

```python
df[df.Dependents == 'No'].sort_values('tenure', ascending=False)
```

### We know how to filter on a value, but what about a list of values — IN condition? In pandas, .isin() operator works the same way. To negate any condition, use ~.

```python
df.InternetService.unique()
```

```python
df[df.InternetService.isin(['DSL', 'Fiber optic'])]
```

```python
# Not In 
df[~df.InternetService.isin(['DSL', 'Fiber optic'])]
```

### Grouping is straightforward: use the .groupby() operator. There’s a subtle difference between semantics of a COUNT in SQL and Pandas. In Pandas, .count() will return the number of non-null/NaN values. To get the same result as the SQL COUNT, use .size().

```python
df.count()
```

```python
type(df.groupby(['gender', 'PhoneService']).size())
```

```python
df.groupby(['gender', 'PhoneService']).size()
```

```python
df.groupby(['gender', 'PhoneService']).size().to_frame('size')
```

```python
df.groupby(['gender', 'PhoneService']).size().to_frame('size').reset_index().sort_values(['PhoneService', 'size'], ascending=[True, False]).reset_index(drop = True)
```
